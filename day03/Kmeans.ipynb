```python
import numpy as np
```

```python
data = np.genfromtxt("solar_system_census.csv",delimiter=',')
```

```python
data=data[1:,1:]
```

```python
data[:5]
```

```python
dist_matrix = np.zeros((data.shape[0], 3))
```

```python
centroids=[[1,100,1],[50,200,1],[58,210,1]]
```

```python
for i in range(3):
    dist_matrix[:, i] = np.linalg.norm(data - centroids[i], axis=1)
```

```python
dist_matrix
```

```python
cluster=np.argmin(dist_matrix, axis=1)
```

```python
data0=data[cluster==0]
```

```python
data[cluster==1]
```

```python
center = [np.mean(data0[:,0]),np.mean(data0[:,1]),np.mean(data0[:,2])]
```
