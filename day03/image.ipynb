```python
class ScrapBooker:
    def crop(array, dimensions, position=(0, 0)):
        if dimensions[0] > array.shape[0] or dimensions[1] > array.shape[1]:
            raise ValueError(
                "The croping dimensions cannot be bigger than the picture's dimensions"
            )
        else:
            return array[position[0] : dimensions[0], position[1] : dimensions[1], :]

```

```python
img2=ScrapBooker.crop(img,(150,200), position=(20,80))
```

```python
    def juxtapose(array, n, axis=0):
        array_jux=np.concatenate((array, array), axis=axis)
        while(n>1):
            array_jux = np.concatenate((array_jux, array), axis=axis)
            n-=1
        return array_jux
```

```python
img4 = juxtapose(img,8,axis=1)
img5 = juxtapose(img4,8,axis=0)
```

```python
plt.imshow(img5)
```

```python
    def mosaic(array, dimensions):
        array_0 = juxtapose(array, dimensions[0], 0)
        return juxtapose(array_0, dimensions[1], 1)
```

```python
    def to_grayscale(array, filter='weighted'):
        if filter == 'm' or filter == 'mean':
            for color_array in array:
                for rgb in color_array:
                    rgb[:] = np.mean(rgb[:])
            return array
        elif filter == 'w' or filter == 'weighted':
            for color_array in array:
                for rgb in color_array:
                    rgb[:] = 0.299 * rgb[0] + 0.587 * rgb[1] + 0.114 * rgb[2]
            return array
```

```python
img11=to_grayscale(img,filter='m')
```
